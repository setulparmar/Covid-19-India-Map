```python
import numpy as np
import pandas as pd
import folium

print('Folium installed and imported!')
```

```python
df=pd.read_csv(r'C:\Users\lenovo\Desktop\state_data.csv')
```

```python
df.rename(columns={'States/UT':'States'},inplace=True)
```

```python
df.head()
```

```python
india_map=folium.Map(location=[20.5937,78.9629],zoom_start=5,tiles='Stamen Toner')
```

```python
incidents = folium.map.FeatureGroup()

# loop through the each state and add each to the incidents feature group
for lat, lng, in zip(df.X, df.Y):
    incidents.add_child(
        folium.vector_layers.CircleMarker(
            [lat, lng],
            radius=4, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
            
        )
    )

# add pop-up text to each marker on the map
latitudes = list(df.X)
longitudes = list(df.Y)
labels = list(df.Confirmed)
labels2=list(df.Recovered)
names=list(df.States)

for lat, lng, label,label2,name in zip(latitudes, longitudes, labels,labels2,names):
    folium.Marker([lat, lng], popup=(str(name),label,label2),tooltip='Click to know Total Confirmed,Recovered Cases').add_to(india_map)    
    
# add incidents to map
india_map.add_child(incidents)
```

```python

india_map.save('Covid19India.html')
```
